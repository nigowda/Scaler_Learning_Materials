```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.randint(low=0, high=100, size=20)
x
```

```python
y = x + np.random.normal(3, 10, len(x))
y
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
plt.vlines(x.mean(), y.min(), y.max(), color='r')
plt.hlines(y.mean(), x.min(), x.max(), color='r')
```

```python
np.corrcoef(x, y)
```

## Salary vs Year of Experience

```python
data = [(1.0, 4), (4.0, 5), (6.0, 7), (9.0, 9), (11.0, 12), (12.0, 15),
        (13.0, 17), (14.0, 18), (15.0, 28), (15.5, 32)]
```

```python
df = pd.DataFrame(data)
df.columns = ['x', 'y']
df
```

```python
plt.scatter(df['x'], df['y'])
```

```python
np.corrcoef(df['x'], df['y'])
```

```python
df['rank_x'] = pd.Series.rank(df.x)
df['rank_y'] = pd.Series.rank(df.y)
df
```

```python
plt.scatter(df['rank_x'], df['rank_y'])
```

```python
np.corrcoef(df['rank_x'], df['rank_y'])
```

```python
from scipy.stats import spearmanr, pearsonr
```

```python
spearmanr(df['x'], df['y'])
```

```python
pearsonr(df['x'], df['y'])
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

spearmanr(x, y)
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [10, 9, 2.5, 6, 4, 3, 2]

pearsonr(x, y)
```

```python
study_hours = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50, 55])
exam_scores = np.array([60, 65, 70, 75, 80, 85, 90, 95, 100, 105])

pearsonr(study_hours, exam_scores)
```
